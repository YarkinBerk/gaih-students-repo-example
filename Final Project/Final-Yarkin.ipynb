```python
# KNOWLEDGE COMP
# 10 question competition 
# Each 1 correct answer 
# Adjust the answers of the questions according to case sensitivity
# Each q 10 points 
# If user got <50 points he/she fails else successful
```

```python
score = 0 # score variable to check the users score

# Here i give the user about the rules of the quiz

print("Welcome the the Best Trivia Quiz of the World! \nHere are the rules! \nIf your score is lower than 50 you fail \nif it is higher that 50 you pass! \nEach question has only one correct answer! \nPlease either pick the correct choice or write the correct answer! \nCareful answers are case sensitive! \n")

# i created each question with if statemends here
# it accepts both letter choice or handwritten answer

answer_1 = input("What is the fastest land animal?  \na. Tiger \nb. Wolf \nc. Giraffe  \nd. Cheetah \nAnswer")

if answer_1 == "d" or answer_1 == "Cheetah":
    score += 10 #if the answer that user given is correct it adds 10 to the score and shows the score if not it doesnt add anything 
    print("Correct!")
    print("Score: ", score)
    print("\n")
else:
    print("Incorrect!")
    print("Score: ", score)
    print("\n")

answer_2 = input("Which country is largest in terms of land size? \na. Turkey \nb. Austria \nc. Croatia \nd. Poland \nAnswer")

if answer_2 == "a" or answer_2 == "Poland":
    score += 10
    print("Correct!")
    print("Score: ", score)
    print("\n")
else:
    print("Incorrect!")
    print("Score: ", score)
    print("\n")
    
answer_3 = input("Which of the following is a real author name and not a pen name? \na. Mark Twain \nb. George Orwell \nc. F. Scott Fitzgerald \nd. Lewis Caroll \nAnswer")

if answer_3 == "c" or answer_3 == "F. Scott Fitzgerald":
    score += 10
    print("Correct!")
    print("Score: ", score)
    print("\n")
else:
    print("Incorrect!")
    print("Score: ", score)
    print("\n")
    
answer_4 = input("Which ist the next number in the Fibonacci sequence: 1,1,2,3,5,8,__? \na. 10 \nb. 11 \nc. 13 \nd. 15 \nAnswer")

if answer_4 == "c" or answer_4 == "13":
    score += 10
    print("Correct!")
    print("Score: ", score)
    print("\n")
else:
    print("Incorrect!")
    print("Score: ", score)
    print("\n")
      
            
answer_5 = input("Which of the following is not a secondary color? \na. Orange \nb. Purple \nc. Yellow \nd. Green \nAnswer")

if answer_5 == "c" or answer_5 == "Yellow":
    score += 10
    print("Correct!")
    print("Score: ", score)
    print("\n")
else:
    print("Incorrect!")
    print("Score: ", score)
    print("\n")
    
answer_6 = input("Which of these is not a movie genre they give an Oscar for? \na. Steaming Movie \nb. Animated Movie \nc. Documentary Movie \nd. International Movie \nAnswer")

if answer_6 == "a" or answer_6 == "Steaming Movie":
    score += 10
    print("Correct!")
    print("Score: ", score)
    print("\n")
else:
    print("Incorrect!")
    print("Score: ", score)
    print("\n")   
    

answer_7 = input("Which of the following is a tilde? \na. # \nb. ~ \nc. & \nd. ^ \nAnswer")

if answer_7 == "b" or answer_7 == "~":
    score += 10
    print("Correct!")
    print("Score: ", score)
    print("\n")
else:
    print("Incorrect!")
    print("Score: ", score)
    print("\n") 

answer_8 = input("How many soccer players should each team have on the field at the start of each macth? \na. 15 \nb. 13 \nc. 9 \nd. 11 \nAnswer")

if answer_8 == "d" or answer_8 == "11":
    score += 10
    print("Correct!")
    print("Score: ", score)
    print("\n")
else:
    print("Incorrect!")
    print("Score: ", score)
    print("\n") 
    
answer_9 = input("What year was the very first model of the iPhone released? \na. 2007 \nb. 2009 \nc. 2004 \nd. 2006 \nAnswer")

if answer_9 == "a" or answer_9 == "2007":
    score += 10
    print("Correct!")
    print("Score: ", score)
    print("\n")
else:
    print("Incorrect!")
    print("Score: ", score)
    print("\n") 
    
answer_10 = input("Which email service is owned my Microsoft? \na. Gmail \nb. Yahoomail \nc. Hotmail \nd. Protonmail \nAnswer")

if answer_10 == "c" or answer_10 == "Hotmail":
    score += 10
    print("Correct!")
    print("Score: ", score)
    print("\n")
else:
    print("Incorrect!")
    print("Score: ", score)
    print("\n") 

# final score

if score <= 49: # if score is lower than 50 it prints "you failed"
    print("Total Score: ", score, "\nYOU FAILED!")
elif score >= 50 and  score < 100: # if score is higher than 50 but not 100 it prints "congratz you passed"
    print("Total Score: ", score, "\nCONGRATZ YOU PASSED!")
elif score == 100: # if the user gets every question right it prints more exiting message to celebrate
    print("Total Score: ", score, "\nWOW YOU GOT ALL OF IT CORRECT \nTIME TO CELEBRATE!!!")
score = 0 # after the quiz i put the score back to 0 otherwise everytime i run the program it will keep adding to the score 
```
